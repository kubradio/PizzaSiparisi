```python
from ast import main
import  csv
import datetime
# menu.txt dosyası olusturulur ve okunacak moda alinir ve ekrana yazdirilacak fonksiyon olusturulur.
def menu_yazdir():
  menu=open("menu.txt","w")
  menu.write("Lütfen Bir Pizza Tabanı Seçiniz:\n1: Klasik\n2: Margarita\n3: TürkPizza\n4: Sade Pizza\nve seçeceğiniz sos:\n11: Zeytin\n12: Mantarlar\n13: Keçi Peyniri\n14: Et\n15: Soğan\n16: Mısır\nTeşekkür ederiz!")
  menu.close()
  menu=open("menu.txt","r")
  print(menu.read())
  menu.close()

# Main fonksiyonu olusturulur
def main():
    menu_yazdir()

menu_yazdir()
#Siparis olusturulmaya baslanan zaman hesaplanir
baslamazamani= datetime.datetime.now()
baslamatarihi=baslamazamani.strftime("%d/%m/%Y %H:%M:%S")
  
# Pizza ust sinifi olusturulur
class pizza:
  def __init__(self, description, cost):
    self.description=description
    self.cost=cost
#Aciklama ve fiyat dondurulur.
  def get_cost(self):
    return self.cost
  def get_description(self):
    return self.description
#Pizza alt sinifi olusturulur

class Klasikpizza(pizza):
  def __init__(self):
    super().__init__("Domates,cheddar,sucuk,mısır,zeytin",40)
class Margaritapizza(pizza):
  def __init__(self):
    super().__init__("Domates,mozarella,fesleğen",30)
class Turkpizza(pizza):
  def __init__(self):
    super().__init__("Domates,kaşar,kavurma,zeytin,sucuk,mısır",50)
class Sadepizza(pizza):
  def __init__(self):
    super().__init__("Domates,kaşar",20)

#Sos ust sinifi olusturulur
class decorator(pizza):
  
  def __init__(self,sos,description,cost):
    super().__init__(description,cost)
    self.sos=sos
#fiyat ve aciklama dondurulur 
  def get_cost(self):
    return self.sos.get_cost()+ super().get_cost()

  def get_description(self):
    return super().get_description() + ' ' + self.sos.get_description()

#sos alt sinifi olusturulur
class zeytinli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "25 gr yeşil zeytin eklenecek",3)
class mantarli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "50 gr mantar eklenecek.", 7)
class kecipeynirli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "20 gr keçi peyniri eklenecek.",7)
class etli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "25 gr kavurma eklenecek.",10)
class soganli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "25 gr soğan eklenecek.",5)
class misirli(decorator):
    def __init__(self, sos):
        super().__init__(sos, "25 gr mısır eklenecek.",3)
   

#Kullaniciya pizza sectirilir
while True:
 
    pizza_secimi = int(input('Pizzanızı seçiniz (1-4): '))

    if pizza_secimi== 1:
        pizza = Klasikpizza()
        break
    elif pizza_secimi== 2:
        pizza = Margaritapizza()
        break
    elif pizza_secimi== 3:
        pizza = Turkpizza()
        break
    elif pizza_secimi== 4:
        pizza = Sadepizza()
        break
    else:
        print("Böyle bir seçenek yok. Tekrar deneyiniz.\n")
        pizza_secimi = int(input('Pizzanızı seçiniz (1-4): '))

#Kullaniciya sos sectirilir
while True:
    
    pizza_sosu = int(input("Sosunuzu seçiniz (11-16): "))

    if pizza_sosu == 11:
        sos = zeytinli(pizza)
        break
    elif pizza_sosu == 12:
        sos = mantarli(pizza)
        break
    elif pizza_sosu == 13:
        sos = kecipeynirli(pizza)
        break
    elif pizza_sosu == 14:
        sos = etli(pizza)
        break
    elif pizza_sosu == 15:
        sos = soganli(pizza)
        break
    elif pizza_sosu == 16:
        sos = misirli(pizza)
        break
    else:   
        print("Böyle bir seçim yok.Tekrar deneyiniz.\n")
        pizza_sosu = int(input("Sosunuzu seçiniz (11-16): "))
# Toplam fiyat hesaplanip ekrana yazdirilir
toplamfiyat = sos.get_cost()


print(f"\nSeçiminiz : {sos.get_description()}\nToplam tutar: {toplamfiyat: .2f} TL\n")


# Musterininismi alinir 
musteriadi = input("\nAdınız: ")
# Tc ve kredi karti bilgisi istenir

   
tc = int(input("TC kimlik numaranız: "))
while True:
            if len(str(tc)) == 11:
               break
            else:
               print("TC kimlik numarası 11 haneli olmalıdır\n")
               tc = int(input("TC kimlik numaranız: "))
while True:
       
            kartno = int(input("Kredi kartı numaranız: "))
            if len(str(kartno)) == 16:
                break
            else:
                print("Kart numarası 16 haneli olmalıdır\n")
                kartno = int(input("Kredi kartı numaranız: "))

            
while True:
    
    kartsifre = input("Kredi kartı şifreniz: ")
    if len(str(kartsifre))==4:
        break
    else:
         print("Kart şifresi 4 haneli olmalıdır\n")
         kartsifre = input("Kredi kartı şifreniz: ")
# Siparisin bittigi zaman ve suresi hesaplanir
bitiszamani=datetime.datetime.now()
gecenzaman=bitiszamani-baslamazamani

# Orders_Database.csv olusturulur
database=open('Orders_Database.csv', 'a') 
databasewrite = csv.writer(database)
databasewrite.writerow(["Adı: ",musteriadi ,"\nTc no: ",tc,"\nPizza türü: " ,sos.get_description(),"\nTutar: ", toplamfiyat, "\nKart no: ",kartno,"\nKart şifresi: " ,kartsifre,"\nİşlem süresi: ",gecenzaman,"\ntarih: ",baslamatarihi])
db=open('Orders_Database.csv', 'a') 
veriler = csv.writer(db)
veriler.writerow(["Adı: ",musteriadi ,"\nTc no: ",tc,"\nPizza türü: " ,sos.get_description(),"\nTutar: ", toplamfiyat, "\nKart no: ",kartno,"\nKart şifresi: " ,kartsifre,"\nİşlem süresi: ",gecenzaman,"\ntarih: ",baslamatarihi])

print(f'\n{sos.get_description()} siparişiniz alınmıştır. Bizi tercih ettiğin için teşekkürler {musteriadi}!')
  



```
